```python
df = pd.read_csv("/home/hoai/user/resource/make-brand/dataset/raw/companies_sorted.csv")
```

```python
df.info()
```

```python
top_brand_countries = [
    'united states', 'united kingdom', 'france', 'germany', 'italy', 
    'sweden', 'switzerland', 'denmark', 'finland', 'netherlands', 
    'singapore', 'australia', 'canada'
]
df = df[df["country"].isin(top_brand_countries)]
df = df[df["name"].str.isascii()]
```

```python
names = [name for name in df["name"] if 3 < len(str(name)) < 20]
import random
name_50 = random.choices(names, k=50000)
with open("dataset/processed/brand_name.txt", 'w', encoding='utf-8') as f:
    f.write("\n".join(name_50))
```
